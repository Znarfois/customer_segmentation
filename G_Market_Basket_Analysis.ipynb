```python
dataset = [['Milk', 'Eggs', 'Bread'], 
           ['Milk', 'Eggs'], 
           ['Milk', 'Bread'], 
           ['Eggs', 'Apple']]
dataset
```

```python
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

te = TransactionEncoder()
te_array = te.fit(dataset).transform(dataset)

df = pd.DataFrame(te_array, columns = te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets_apriori = apriori(df, min_support = 0.01, use_colnames = True)
```

```python
frequent_itemsets_apriori
```

```python
from mlxtend.frequent_patterns import fpgrowth

frequent_itemsets_fp = fpgrowth(df, min_support = 0.01, use_colnames= True)
```

```python
frequent_itemsets_fp
```

```python
from mlxtend.frequent_patterns import association_rules

#min_threshold is being x% sure that buying something will also mean buying something else
#arbitrary number based on your experience and domain knowledge
rules_ap = association_rules(frequent_itemsets_apriori, metric='confidence', min_threshold = 0.8)
rules_fp = association_rules(frequent_itemsets_fp, metric='confidence', min_threshold = 0.8)
```
