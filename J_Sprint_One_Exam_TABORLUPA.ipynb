### 2) FizzBuzz

Write a short program that prints each number from 1 to 100 on a new line. 

- For each multiple of 3, print "Fizz" instead of the number. 
- For each multiple of 5, print "Buzz" instead of the number. 
- For numbers which are multiples of both 3 and 5, print "FizzBuzz" instead of the number.

```python
for i in range(1, 101):
    if ((i % 5 == 0) and (i % 3 == 0)): 
        print('FizzBuzz')
    elif (i % 3 == 0):
        print('Fizz')
    elif (i % 5 == 0):
        print('Buzz')
    else:
        print(i)
```

### 3) Demographia

Using `countries.csv` in S3 bucket:

- Find the minimum and maximum of the 'life expectancy' for costal and landlocked countries.
- How many coastal and landlocked countries are in each continent? Determine the mean deathrate for coastal and landlocked countries in each continent.
- Create a new column measuring population density.  Determine the mean population density of all the countries.  Create a new column 'dense' that is True for countries with higher than average density and False for lower than average density.
- For each category of 'dense' compute the mean birthrate and the difference between the maximal birthrate and the minimal birthrate.

```python
import pandas as pd

df = pd.read_csv("countries.csv", delimiter = ";")
df.head()
df['Population'].max()
```

```python
df.shape
```

```python
df_coastal = df[df['Coastline']>0]
df_coastal.head()
```

```python
df_landlocked = df[df['Coastline']==0]
df_landlocked.head()
```

```python
# Find the minimum and maximum of the 'life expectancy' for costal and landlocked countries.
```

```python
min(df_coastal['Life expectancy'])
```

```python
max(df_coastal['Life expectancy'])
```

```python
min(df_landlocked['Life expectancy'])
```

```python
max(df_landlocked['Life expectancy'])
```

```python
# How many coastal and landlocked countries are in each continent? 
```

```python
df_coastal.groupby('Continent').agg({'Country': 'count'})
```

```python
df_landlocked.groupby('Continent').agg({'Country': 'count'})
```

```python
#Determine the mean deathrate for coastal and landlocked countries in each continent.

df_coastal.groupby('Continent').agg({'Country': 'count',
                                    'Deathrate': lambda x: x.sum() / x.count()})
```

```python
df_landlocked.groupby('Continent').agg({'Country': 'count',
                                    'Deathrate': lambda x: x.sum() / x.count()})
```

```python
# Create a new column measuring population density. Determine the mean population density of all the countries. 

df['Population Density'] = df['Population'] / df['Area']
df[['Country','Population Density']]
```

```python
import numpy as np

average_density = np.mean(list(df[df['Area'] != 0]['Population Density']))
```

```python
average_density
```

```python
#Create a new column 'dense' that is True for countries with higher than average density and False for lower than average density

#average_density = 25
df['dense'] = df['Population Density'] > average_density
df[['Population Density', 'dense']]
```

```python
df.groupby('Birthrate').agg({'Birthrate': 'mean',
                                'Deathrate': lambda x: x.sum() / x.count()})
```

```python
# For each category of 'dense' compute the mean birthrate and the difference between the maximal birthrate and the minimal birthrate.

mean_birthrate = df['Birthrate'].mean()
mean_birthrate
df
```

```python
df[(df['dense']==True) & (df['Area'] != 0)]['Population Density'].mean()
```

```python
df[(df['dense']==False) & (df['Area'] != 0)]['Population Density'].mean()
```

```python
birthrate_diff = max(df['Birthrate']) - min(df['Birthrate'])
birthrate_diff
```
